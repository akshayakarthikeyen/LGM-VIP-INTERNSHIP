# image to pencil sketch

# importing libraries

```python
import cv2
from PIL import Image
import matplotlib.pyplot as plt
```

```python
image_location='C:/Users/aksha/Downloads/'
imagename = 'eye.jpg'
```

# read the image

```python
img = cv2.imread(image_location+imagename)
```

# display the image

```python
img = cv2.resize(img,(500,300))
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10,10))
plt.imshow(img)
```

# convert image to gray scale image

```python
grey_image=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(grey_image)
```

# invert the image

```python
inverted_grey_image=255-grey_image
plt.figure(figsize=(10,10))
plt.imshow(inverted_grey_image)
```

# blur the image by gaussian function

```python
blurred_image=cv2.GaussianBlur(inverted_grey_image,(21,21),0)
plt.figure(figsize=(10,10))
plt.imshow(blurred_image)
```

# invert the blurred image

```python
inverted_blurred_image=255-blurred_image
plt.figure(figsize=(10,10))
plt.imshow(inverted_blurred_image)
```

# create the pencil sketch

```python
pencil_sketch = cv2.divide(grey_image,inverted_blurred_image,scale=256.0)
cv2.imwrite("outputImage.png", pencil_sketch)
plt.figure(figsize=(10,10))

display('Pencil sketch', Image.fromarray(pencil_sketch))
```

```python
display('original Image', Image.fromarray(img))
display('Pencil sketch', Image.fromarray(pencil_sketch))
```
